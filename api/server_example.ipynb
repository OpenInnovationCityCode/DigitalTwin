```python
import requests

endpoint_place = "http://127.0.0.1:5000/api/place"
data_place = {'name':'tree','long':49,'lat':31}
```

```python
r1 = requests.post(endpoint_place,json=data_place)
print(r1.text)
```

```python
endpoint_delete = "http://127.0.0.1:5000/api/delete"
data_delete = {'id':12}

r2 = requests.post(endpoint_delete,json=data_delete)
print(r2.text)
```

```python
endpoint_get_world = "http://127.0.0.1:5000/api/get_world"
r3 = requests.get(endpoint_get_world)
print(r3.text)
```
